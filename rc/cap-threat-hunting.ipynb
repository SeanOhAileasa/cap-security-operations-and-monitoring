###### Proactive Threat Hunting
***

Threat Hunting is a cybersecurity technique designed to detect the presence of threats that have not been discovered by normal security monitoring.

When performing proactive threat hunting, the ultimate goal is to either locate a legitimate threat that was unbeknownst beforehand or create a new detection rule.

Some common techniques that are practiced with threat hunting include:

- Establishing a hypothesis to help start and guide your search.


- Effectively using threat intelligence to profile your threat actors by their activities.


- Analyzing network traffic for possible activity to malicious domains.


- Looking for suspicious processes attempting to operate outside of their means.


- Comparing previously infected hosts for similarities of behavioral trends.


- Blocking known threat vectors as they arise.

###### Methodology - ``SMART``
***

When considering time spent on proactive threat hunting, it will be important to use the **SMART** methodology to achieve a maximum benefit from your resources allocated to this effort:

- Specific


- Measurable


- Achievable


- Realistic


- Timely

###### Benefits
***

Some examples of benefits gained from proactive threat hunting include:

- Improving your overall detection capabilities and visibility


- Better using your threat intelligence


- Reducing overall attack surface and risk exposure


- Learning which assets can be bundled, or in other words, the bulk is updated to remove widespread vulnerabilities
