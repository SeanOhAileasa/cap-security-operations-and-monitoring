###### Network Forensics Analysis Tools
***

The following network forensics analysis tools will help with security monitoring:

> Switched Port Analyzer (``SPAN``)

Dedicated port on a switch that is a mirrored copy of the network traffic passing by.

> Test Access Port (``TAP``)

A hardware device inserted into a cable to copy frames for analysis.

> Packet Sniffer

Records data from frames as they pass over network media, using methods such as a mirror port or tap device.

###### Network Forensics Analysis Tools - ``NFAT``
***

The following tools either use or can be used to detect packet sniffing activities on a network.

> ``tcpdump``

Command-line packet sniffing utility - syntax: 

```
tcpdump -i eth
```

This is the base syntax, where ``eth`` represents the interface you are trying to listen on - common switch commands [ [``Options``](https://www.tcpdump.org/manpages/tcpdump.1.html) ]:

- ``-i`` - Interface.


- ``-X`` - Capture the payload in hex and ASCII.


- ``-XX`` - Capture in hex and ASCII and include the data link header.


- ``-n`` - Show address in numeric format.


- ``-e`` - Include data link header.


- ``-s`` - Set a custom bytes length.


- ``-w`` - Write the output to a file.


- ``-v`` (``-vv`` or ``-vvv``) - Increase the verbosity.

> Wireshark

Wireshark is similar to ``tcpdump`` - can use the same syntax commands in Wireshark. 

Wireshark also has a GUI interface that lays out all of its intuitive options in a Microsoft-like interface. 

If you prefer just to have the command-line version - download ``tshark``.

###### IP and DNS Analysis
***

> Domain Generation Analysis

Known-bad IP addresses appear on one of more blacklists. 

Gathered from static malware with traditional ``C2`` communications.

Can be looked up using one of many online reputation resources to determine if they are indeed nefarious - the following are some examples of these services:

- [ [``brightcloud.com``](https://www.brightcloud.com) ]


- [ [``mxtoolbox.com``](https://www.mxtoolbox.com) ]


- [ [``urlvoid.com``](https://www.urlvoid.com) ]


- [ [``ipvoid.com``](https://www.ipvoid.com) ]

> Evade Blacklists - Domain Generation Algorithm (``DGA``)

Method used by malware to evade blacklists by generating domain names for ``C&C`` networks dynamically.

``DGA`` operates in the following steps:

1. Attacker sets up fake dynamic DNS service using fraudulent credentials and payments methods to cover their tracks.


2. Malware executes to create a list of new domains (hiding underneath a legitimate domain).


3. Another parallel ``DGA`` generates a list of Dynamic DNS (``DDNS`` records of possible names for it to use.


4. When a ``C2`` connection needs to be made, it chooses a domain from the list it created.


5. Each time a new ``C2`` connection is made, a new domain is generated, thus frustrating any defending attempts to block attacks by domain.

###### Denial of Service Mitigation - Blackholes and Sinkholes
***

- ``Blackhole`` is a means of mitigating ``DoS`` or intrusion attacks by silently dropping traffic. 


- ``Sinkhole`` is a ``DoS`` attack mitigation strategy that directs the traffic that is flooding a target IP address to a different network for analysis.

The main difference to remember between the two is that a ``Blackhole`` is where logs go to die, whereas with a ``Sinkhole``, the logs are sent to an isolated location for further monitoring.

###### Proxy Log Review
***

Reviewing ``Proxy Logs`` can help to determine the true intent behind HTTP requests.

There are several different types of proxies and their specific functions to be aware of:

> Forward Proxy

A server that forwards requests from a private network to the Internet via a firewall.

> Nontransparent Proxy

A server that redirects requests and responses for clients configured with the proxy address and port.

> Transparent Proxy

A server that redirects requests and responses without the client being configured to use it - also referred to as a ``Forced`` or ``Intercepted`` proxy.

> Reverse Proxy

A server that protects servers from direct contact with client requests.

###### Anomalous Behavior Analysis
***

Conducting anomalous behavior analysis.

> Known-Good Processes

Shellcode is easy to obfuscate - can often evade signature-based ``AV`` products. 

Threat hunting and security monitoring must use behavioral-based techniques to identify infections - this means close analysis of the processes running in system memory on a host. 

To perform process analysis effectively, you should build up a sense of what is ``"normal"`` in a system and spot deviations in a potentially infected system. 

Also need to use appropriate analysis tools. 

- [ [``Sysinternals``](https://docs.microsoft.com/en-us/sysinternals) ] is a suite of tools designed to assist with troubleshooting issues with Windows, and many of the tools are suited to investigating security issues.

When hunting for a malicious process using a tool such as ``Process Explorer`` (part of ``Sysinternals``), you need to be able to filter out the legitimate activity known-good behavior — generated by normal operation of the computer and look for the signs of anomalous behavior that could identify a process as suspicious.

- To identify what might be suspicious on a typical Windows host, you must understand how legitimate processes are structured. 

The following represent the main processes launched by the ``SYSTEM`` account:

- System Idle (``PID 0``) and System (``PID 4``) — Kernel-level binaries - will generate CPU and network activity - System is the parent for ``Interrupts`` and the ``Windows Session Manager SubSystem`` (``smss.exe``) - ``SMSS`` is the first user-mode process and should only appear as a child of System and launch from ``SystemRoot%\System32``.


- ``Client Server Runtime SubSystem`` (``csrss.exe``) — manages low-level Windows functions - is normal for there to be several running (so long as launched from ``%SystemRoot%\System32``) and for them to have no parent.


- ``WININIT`` (``wininit.exe``) — manages drivers and services - should only be one instance of ``WININIT``.


- ``services.exe`` — Hosts nonboot drivers and background services - there should only be one instance of ``services.exe``, running as a child of ``wininit.exe`` - each service process should either appear as a child process of ``services.exe`` or as a child process of a ``svchost.exe`` wrapper (which should always load from ``%SystemRoot%\System32``).

Check running services to ensure that they are digitally signed (show the Verified Signer column and run ``Options > Verify Image Signatures``). Services will be started by the ``SYSTEM``, ``LOCAL SERVICE``, or ``NETWORK SERVICE`` accounts.


- ``Local Security Authority SubSystem`` (``lsass.exe``) — handles authentication and authorization - there should only be a single instance, running as a child of ``wininit.exe``.


- ``WINLOGON`` (``winlogon.exe``) - manages access to the user desktop - there will be one instance for each user session - the ``Desktop Window Manager`` (``dwm.exe``) is likely to be a child process in modern versions of Windows.


- ``USERINIT`` (``userinit.exe``) — sets up the shell (typically ``explorer.exe``) and then quits - should only see this process briefly after log-on.


- ``Explorer`` (``explorer.exe``) — this is the typical user shell, launched with the user's account privileges rather than SYSTEM's, and is likely to be the parent for processes started by the logged-on user.

Should have some idea what system-level processes are and how they execute to be able to properly identify when a malicious process is attempting to masquerade.

Given the potential exploit techniques, to locate a malicious process you may be looking for a process name that you do not recognize, or for a valid process name that
is not entirely as it should be in other respects - refer ``Official CompTIA CySA+ Student Guide - p.101/102``

###### Digital Signatures
***

Think of ``Digital Signatures``, think ``S/MIME`` (Secure / Multipurpose Internet Mail Extensions). 

``S/MIME`` is the method of Digital Signature exchange that facilitates email message delivery and security.

The ``S/MIME`` exchange goes as follows:

1. Drew sends Kenny his Digital Certificate that contains his ``Public Key`` and ``ID`` (signing the message using his ``Private Key``).


2. Kenny uses his ``Public Key`` to decode the signature from the message to validate Drew's ``ID`` and determine that the message is legitimate.


3. Kenny responds with his ``Public Key`` and ``Digital Certificate`` - then, Drew does the same process to trust Kenny.


4. When neither party wants to encrypt an email, the same process applies - the sending party then uses their ``Private Key`` to sign the encrypted message, and the receiving party uses theirs to decode it.
