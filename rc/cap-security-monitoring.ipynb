###### Network Forensics Analysis Tools
***

The following network forensics analysis tools will help with security monitoring:

> Switched Port Analyzer (``SPAN``)

Dedicated port on a switch that is a mirrored copy of the network traffic passing by.

> Test Access Port (``TAP``)

A hardware device inserted into a cable to copy frames for analysis.

> Packet Sniffer

Records data from frames as they pass over network media, using methods such as a mirror port or tap device.

###### Network Forensics Analysis Tools - ``NFAT``
***

The following tools either use or can be used to detect packet sniffing activities on a network.

> ``tcpdump``

Command-line packet sniffing utility - syntax: 

```
tcpdump -i eth
```

This is the base syntax, where ``eth`` represents the interface you are trying to listen on - common switch commands [ [``Options``](https://www.tcpdump.org/manpages/tcpdump.1.html) ]:

- ``-i`` - Interface.


- ``-X`` - Capture the payload in hex and ASCII.


- ``-XX`` - Capture in hex and ASCII and include the data link header.


- ``-n`` - Show address in numeric format.


- ``-e`` - Include data link header.


- ``-s`` - Set a custom bytes length.


- ``-w`` - Write the output to a file.


- ``-v`` (``-vv`` or ``-vvv``) - Increase the verbosity.

> Wireshark

Wireshark is similar to ``tcpdump`` - can use the same syntax commands in Wireshark. 

Wireshark also has a GUI interface that lays out all of its intuitive options in a Microsoft-like interface. 

If you prefer just to have the command-line version - download ``tshark``.

###### IP and DNS Analysis
***

> Domain Generation Analysis

Known-bad IP addresses appear on one of more blacklists. 

Gathered from static malware with traditional ``C2`` communications.

Can be looked up using one of many online reputation resources to determine if they are indeed nefarious - the following are some examples of these services:

- [ [``brightcloud.com``](https://www.brightcloud.com) ]


- [ [``mxtoolbox.com``](https://www.mxtoolbox.com) ]


- [ [``urlvoid.com``](https://www.urlvoid.com) ]


- [ [``ipvoid.com``](https://www.ipvoid.com) ]

> Evade Blacklists - Domain Generation Algorithm (``DGA``)

Method used by malware to evade blacklists by generating domain names for ``C&C`` networks dynamically.

``DGA`` operates in the following steps:

1. Attacker sets up fake dynamic DNS service using fraudulent credentials and payments methods to cover their tracks.


2. Malware executes to create a list of new domains (hiding underneath a legitimate domain).


3. Another parallel ``DGA`` generates a list of Dynamic DNS (``DDNS`` records of possible names for it to use.


4. When a ``C2`` connection needs to be made, it chooses a domain from the list it created.


5. Each time a new ``C2`` connection is made, a new domain is generated, thus frustrating any defending attempts to block attacks by domain.

###### Denial of Service Mitigation - Blackholes and Sinkholes
***

- ``Blackhole`` is a means of mitigating ``DoS`` or intrusion attacks by silently dropping traffic. 


- ``Sinkhole`` is a ``DoS`` attack mitigation strategy that directs the traffic that is flooding a target IP address to a different network for analysis.

The main difference to remember between the two is that a ``Blackhole`` is where logs go to die, whereas with a ``Sinkhole``, the logs are sent to an isolated location for further monitoring.

###### Proxy Log Review
***

Reviewing ``Proxy Logs`` can help to determine the true intent behind HTTP requests.

There are several different types of proxies and their specific functions to be aware of:

> Forward Proxy

A server that forwards requests from a private network to the Internet via a firewall.

> Nontransparent Proxy

A server that redirects requests and responses for clients configured with the proxy address and port.

> Transparent Proxy

A server that redirects requests and responses without the client being configured to use it - also referred to as a ``Forced`` or ``Intercepted`` proxy.

> Reverse Proxy

A server that protects servers from direct contact with client requests.
